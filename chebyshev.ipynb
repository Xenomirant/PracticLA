```python
U = np.random.uniform(-1, 1, 50000)
```

```python
x = np.linspace(-.1, 2.1, num = 5000)
```

```python
f_x = lambda x: x**3 + x**2 + x + 1
```

```python
p_x = lambda x: x**2 + 7/4*x + 1
```

```python
q_x = lambda x: 4*x**2 - 5/4*x + 5/4
```

```python
plt.plot(x, f_x(x), label = "f_x")
plt.plot(x, p_x(x), label = "p_x")
plt.plot(x, q_x(x), label = "q_x")
plt.legend()
```

```python
plt.plot(x, f_x(x) - p_x(x), label = "f_x(x) - p_x(x)")
plt.plot(x, f_x(x) - q_x(x), label = "f_x(x) - q_x(x)")
plt.legend()
```
