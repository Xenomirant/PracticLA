```python
f = scp.interpolate.interp1d(x, y, kind="cubic", fill_value="extrapolate")
```

```python
xnew = np.arange(-2.3, 3, 0.1)

ynew = f(xnew)   # use interpolation function returned by `interp1d`

plt.plot(x, y, 'o', xnew, ynew, '-')

plt.show()
```

```python
def approx_sin(range_pi = 10):

    x = np.arange(0, 2*range_pi*np.pi, np.pi/32)

    breaks = np.split(x, range_pi*2)

    start_matr = np.array([[1, 0, 0],
                          [0, 1, 0],
                          [1, np.pi/2, np.pi**2/4]])

    start_y = np.array([0, 1, 0])

    coefs = scp.linalg.inv(start_matr) @ start_y

    spline = []

    f = lambda x: coefs[2]*x**2 + coefs[1]*x + coefs[0]

    spline.append(f(breaks[0]))
    
    for i in range(1, len(breaks)):

        new_matr = np.array([[1, np.pi/2*i, (np.pi/2*i)**2],
                            [0, 1, i*np.pi],
                            [1, (i+1)*np.pi/2, ((i+1)*np.pi/2)**2]])
        print(new_matr)

        new_y = np.array([i % 2, i*np.pi*coefs[2]*np.pi+coefs[1], (i+1)%2])
        print(new_y)

        coefs = scp.linalg.inv(new_matr) @ new_y

        f = lambda x: coefs[2]*x**2 + coefs[1]*x + coefs[0]

        spline.append(f(breaks[i]))



    return breaks, spline
```

### Bezier

```python
Points = np.array([[-1, -1], 
                  [1, 3],
                  [2, 4],
                  [7, -1]])

bez = []

for i in range(0, 1000):

    x = i/1000
    
    m = np.array([(1-x)**3, (1-x)**2*x, (1-x)*x**2, x**3])

    bez.append(m@Points)
```

```python
#plt.plot([item[0] for item in bez], [item[1] for item in bez])
plt.plot(*zip(*Points))
plt.plot(*zip(*bez))
plt.show()
```
